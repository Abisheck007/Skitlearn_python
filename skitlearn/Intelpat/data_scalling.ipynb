```python
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import fetch_openml
df = fetch_openml('titanic',as_frame=True,version=1)['data']
df
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
df['sex'] = enc.fit_transform(df[['sex']]).toarray()
df
```

```python
#to work with scalling we need to know what are the int values present in our dataset
nums= df.select_dtypes(include=['int64','int32','float64']).columns
nums
```

```python
#now we do scalling here
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
df[nums] = ss.fit_transform(df[nums])
df[nums].describe()
df[nums]
```
