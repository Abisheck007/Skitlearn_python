```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
```

```python
titanic = fetch_openml('titanic',version=1,as_frame=True)
df = titanic['data']
df
```

```python
df['survived'] = titanic['target']
df
```

```python
sns.countplot(x='survived',hue='sex',data=df)
```

```python
#same like sex we can also hue with pclass to know from which class more people are died and survived!
sns.countplot(data=df,x = 'survived',hue='pclass')
```

```python
df['family'] = df['sibsp']+df['parch']
df.loc[df['family']>0,'travelling alone'] = 0
df.loc[df['family']==0,'travelling alone'] = 1
df.drop(['sibsp','parch'],axis=1,inplace=True)
df
```

```python
df.drop(['name','ticket','home.dest'],axis=1,inplace=True)
df
```

```python
df.drop(['cabin','body','boat'],axis=1,inplace=True)
df
```

```python
#now we need to change the value of sex into int so we can use two techniques one is onehotencoder and dummies,,,we already seen onehot so now we use dummies
sex  = pd.get_dummies(df['sex'],drop_first=True)
sex
df['sex'] = sex
```

```python
df
```

```python
df['sex'] = df['sex'].replace({'False':0,'True':1})
df['sex'] = df['sex'].astype(int)

```

```python
df.isnull().sum()
```

```python
#to remove null values we use value imputation technique
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy='mean')
```

```python
df['age'] = si.fit_transform(df[['age']])
df['fare']=si.fit_transform(df[['fare']])
```

```python
sim = SimpleImputer(strategy='most_frequent')
df['embarked'] = sim.fit_transform(df[['embarked']]).ravel()
```

```python
embark = pd.get_dummies(df['embarked'])
```

```python
a1 = embark.astype(int)
a1
```

```python
df = pd.concat([df,a1],axis=1)
df
```

```python
df.drop(['C','S','Q'],axis=1)
```

```python
df
```

```python
df = pd.concat([df,a1],axis=1)
```

```python
df = df.drop(['fare'],axis=1)
```

```python
#now the dataset is good to do logistic regression
#we take survived as y and other as x here y is label and x is feature
x = df.drop(['survived'],axis=1)
x
```

```python
y = df['survived']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print(xtrain.shape,ytrain.shape)
print(xtest.shape,ytest.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(xtrain,ytrain)
prediction = model.predict(xtest)
accuracy = accuracy_score(ytest,prediction)
print("prediction",prediction)
print("accuracy:",accuracy)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,prediction)
```
