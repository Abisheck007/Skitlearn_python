```python
import pandas as pd    
df = pd.read_csv("health_data_200.csv")
df
```

```python
df.info()
```

```python
#use label encoder for hear left and right
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['hearing(left)'] = le.fit_transform(df['hearing(left)'])
```

```python
df['hearing(right)'] = le.fit_transform(df['hearing(right)'])
df
```

```python
x = df.drop(['smoking'],axis=1)
y = df['smoking']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
model = RandomForestClassifier()
model.fit(xtrain,ytrain)
predicton = model.predict(xtest)
print(predicton)
accuracy = accuracy_score(ytest,predicton)
print(accuracy)
```

```python
#now we use hyper parameters tuning using gridsearch technique
from sklearn.model_selection import GridSearchCV
forest_params = [{'max_depth':list(range(10,15)),'max_features':list(range(1,14))}]
gs = GridSearchCV(model,forest_params,cv = 10,scoring='accuracy')
gs.fit(xtrain,ytrain)
gs.best_params_
gs.best_score_
```

```python
predicton1 = gs.predict(xtest)
print(predicton1)
acc = accuracy_score(ytest,predicton1)
print(acc)
```
