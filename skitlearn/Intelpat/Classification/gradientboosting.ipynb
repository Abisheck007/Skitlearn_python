```python
from sklearn import datasets
data = datasets.load_diabetes()
data
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data=data.data,columns=data.feature_names)
df['target'] = data.target
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
x = df.drop(['target'],axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=0)

```

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier(n_estimators=500)
model.fit(xtrain,ytrain)
prediction  = model.predict(xtest)
prediction
```

```python
from sklearn.metrics import mean_absolute_error
diff_change = mean_absolute_error(ytest,prediction)
diff_change
```

```python
#you see above you see that the error rate is so high we need to do parameter tuning
```

```python
feature_scores = pd.Series(model.feature_importances_,index=xtrain.columns).sort_values(ascending=False)
feature_scores
```

```python
#we will see that which data contribute more for our model  and which is less
sns.barplot(x = feature_scores,y=feature_scores.index)
```

```python
#you see in the graph sex and s4 doesnt contribute much for our model so we can drop them and see the error rate
x1 = df.drop(['s4','target','sex'],axis=1)
y1 = df['target']
```

```python
x1train,x1test,y1train,y1test = train_test_split(x1,y1,test_size=0.2)
model1 = GradientBoostingClassifier(n_estimators=500)
model1.fit(x1train,y1train)
```

```python
pred1 = model1.predict(x1test)
pred1
```

```python
error= mean_absolute_error(y1test,pred1)
error
```
