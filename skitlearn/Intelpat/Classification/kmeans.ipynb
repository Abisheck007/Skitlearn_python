```python
 import pandas as pd
 import numpy as np
 from sklearn import datasets
```

```python
data =datasets.load_wine()
data
```

```python
x = data.data
y = data.target
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x1 = ss.fit_transform(x)
```

```python
#now the clustering part starts with elbow technique
#now we need to loop the dataset to know the cluster value to fit
from sklearn.cluster import KMeans
wss =[]
for i in range(1,11):
    model = KMeans(n_clusters=i,init='k-means++',random_state=0)
    model.fit(x1)
    wss.append(model.inertia_)
    
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wss)
plt.title("the Elbow Technique")
plt.xlabel("no of cluster")
plt.ylabel("wss")
plt.show()
```

```python
#now we find that the elbow is form at 3 so the k value is 3(n_cluster)
#now instead instead of putting cluster value as i we can put 3 and try to get the prediction
model = KMeans(init='k-means++',n_clusters=3,n_init=10,max_iter=360)
model.fit(x1)
labels = model.labels_#why we are giving label instead od prediction means it is a unsupervised algorithm it dont even have a label or target column so finding the label is our motto.
labels
```

```python
#now we find the label so we check the accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(labels,y)
accuracy
```
