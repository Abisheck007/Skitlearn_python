```python
from sklearn.datasets import fetch_openml
```

```python
df = fetch_openml('titanic',version=1,as_frame=True)['data']
df
```

```python
df.info()
df.isnull().sum()
```

```python
#by using this value imputation we can able to fill the missing values let see how:
from sklearn.impute import SimpleImputer
print("the values in the age column before computation:" , df.age.isnull().sum())
se = SimpleImputer(strategy='mean')
df['age'] = se.fit_transform(df[['age']])
print("the values in the age column after computation:" , df.age.isnull().sum())
```

```python
df
```

```python
#here we combined sibsp and parch into one and find who are travelled alone
df['family'] = df['sibsp'] + df['parch']
df.loc[df['family']>0,'travelled_alone'] = 0#see the dataset if the value in sibsp and parch is 0 means then he is travelling alone because he has no siblings(sibsp) and no parent children bond(parch)
df.loc[df['family']==0,'travelled_alone'] = 1 
df['travelled_alone'].value_counts().plot(kind='bar',title='passangers who travelled alone')
```
